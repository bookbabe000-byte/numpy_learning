```python
import numpy as np
```

```python
## concatenation 

arr1=np.array([1,2,3,4])
arr2=np.array([4,6,7,8])
```

```python
arr1,arr2
```

```python
arr3=np.concatenate((arr1,arr2))
```

```python
arr3
```

```python
arr3=np.concatenate((arr1,arr2),axis=0)
```

```python
arr3
```

```python
arr5=np.array([[1,2,3],[4,5,6]])
arr6=np.array([[7,8,9],[10,11,12]])

```

```python
arr=np.concatenate((arr5,arr6),axis=1)
```

```python
print(arr)
```

```python
arr=np.vstack((arr5,arr6))
```

```python
print(arr)
```

```python
arr=np.hstack((arr5,arr6))
print(arr)
```

```python
arr=np.dstack((arr5,arr6))
print(arr)
```

```python
array7=np.array([[1,2,3],
                [4,5,6]])
array8=np.array([[2,2,4]])  
arr9=np.vstack((array7,array8))
print(arr9)              
```
